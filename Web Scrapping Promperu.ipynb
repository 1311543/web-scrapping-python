```python
# FUNCION PARA ENVIAR LISTA LA DATA EN ARRAY
def prepare_data(section):
    prepare_section = section.findAll("option")
    prepared_data = []
    for i in range(len(prepare_section)):
        prepared_data.append(prepare_section[i].attrs['value'])
    return prepared_data[0:]
```

```python
def wait_element(element, time, driver_page):
    try :
        WebDriverWait(driver_page, time).until(EC.presence_of_element_located((By.XPATH, element)))
        print(time)
        return [True if EC.presence_of_element_located((By.XPATH, element)) else False] 
    
    except TimeoutException:
        return print("Timed out waiting for page to load")
    #except NoSuchElementException:
     #   return False
```

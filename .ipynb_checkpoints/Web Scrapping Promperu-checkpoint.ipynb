```python
url = 'http://www.siicex.gob.pe/promperustat/frmBuscar_Pais.aspx'
page = requests.get(url)
```

```python
page.content
soup = BeautifulSoup(markup=page.content, features='html.parser')
soup
```

```python
soup.prettify
```

```python
#section_ = soup.find(name='ul', attrs={'id': 'avisos-content'})
section_pais = soup.find(name='select', id='ctl00_ContentPlaceHolder1_ddlPais')
section_año = soup.find(name='select', id='ctl00_ContentPlaceHolder1_ddlAnioI')
section_mes = soup.find(name='select', id='ctl00_ContentPlaceHolder1_ddlMesI')
section_fob = soup.find(name='select', id='ctl00_ContentPlaceHolder1_ddlOrdenadoPor')
```

```python
section_año
```

```python
# FUNCION PARA ENVIAR LISTA LA DATA EN ARRAY
def prepare_data(section):
    prepare_section = section.findAll("option")
    prepared_data = []
    for i in range(len(prepare_section)):
        prepared_data.append(prepare_section[i].attrs['value'])
    return prepared_data[0:]
```

```python
paises = prepare_data(section_pais)
año = prepare_data(section_año)
mes = prepare_data(section_mes)
fob = prepare_data(section_fob)
```

```python
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.maximize_window()
driver.get("http://www.siicex.gob.pe/promperustat/frmBuscar_Pais.aspx")
assert "PROMPERUSTAT - Estadísticas de Exportaciones Peruanas" in driver.title
consultar = driver.find_element_by_id("ctl00_ContentPlaceHolder1_btConsultar")

element_pais = driver.find_element_by_xpath("//select[@id='ctl00_ContentPlaceHolder1_ddlPais']")
element_año = driver.find_element_by_xpath("//select[@id='ctl00_ContentPlaceHolder1_ddlAnioI']")

all_options_country = element_pais.find_elements_by_tag_name("option")

print(all_options_country)

for option in all_options_country:
    print("Value is: %s" % option.get_attribute("value"))
    option.click()

consultar.click()
driver.quit()
```
